```python
#--------------------------------------------------libraries------------------------------------#
import numpy as np  
import matplotlib.pyplot as plt 
import pandas as pd
from pandas import DataFrame
import csv
import random 
import seaborn as sns 
class struct: 
   
    
    def __init__(self,data, height): 
        self.data = data
 
        self.ptr = [None]*(height+1) 
        
  
class List: 
    
    def __init__(self, limmit, n): 
        
        self.cut = limmit 
        self.n = n
        self.next = self.addStruct(self.cut, -1) 
        self.height = 0
    
    def headAndTail(self):
        noOfLevel = 0
        while random.randint(0, 1) ==1 and noOfLevel<self.cut:
            noOfLevel += 1
        return noOfLevel
            
    def randomLevel(self): 
        lvl = 0
    
        while random.random()<self.n and lvl<self.cut:
            lvl += 1
        
        return lvl  
    
    def addStruct(self, lvl, data): 
        n = struct(data, lvl) 
        return n 
      

    
    def print_list(self): 
        
        head = self.next
        for lvl in range(self.height+1): 
            print("Level {}: ".format(lvl), end=" ") 
            struct = head.ptr[lvl] 
            while(struct != None): 
                print(struct.data, end=" ") 
                struct = struct.ptr[lvl] 
            print("") 
    
    def Insertion(self, data): 
        temp_level=0

        temp = [None]*(self.cut+1) 
        
        temp_ptr = self.next
        #print("see ya :",temp_ptr.ptr)
        
        for i in range(self.height, -1, -1): 
            #print("self.height is :",self.height)
            #print("i is :",i)
            while temp_ptr.ptr[i] and temp_ptr.ptr[i].data < data:
                #print("I came here in while loop with i :",i)
                #print("current.ptr[i]  is :",temp_ptr.ptr[i] )
                #print("current.ptr[i].data  is :",temp_ptr.ptr[i].data )
                temp_ptr = temp_ptr.ptr[i] 
                
            #print("hehe bro for :",data)
            temp[i] = temp_ptr 
            #print("update is :",temp[i].ptr)
  
     
        temp_ptr = temp_ptr.ptr[0] 
        #print("val of current is :",temp_ptr)
        #print("good joke")
        #print("current.key :",current.key)
        #print("current.forward :",current.forward)
        
        #if temp_ptr != None and temp_ptr.data == data:
            #print("This Element already exists in the skiplist,so no need of redundency")
        if temp_ptr == None or temp_ptr.data != data: 
            # Generate a random level for node 
            #print("I am in if block")
            rlevel = self.randomLevel() 
            #rlevel=self.headAndTail()
            
            temp_level=rlevel
            
            if rlevel > self.height: 
                #print("my rlevel is greater than self height :")
                for i in range(self.height+1, rlevel+1): 
                    #print("update[i] before assigning header is :",temp[i])
                    #print("update[i].header before assigning header is :",update[i].header)
                    temp[i] = self.next
                    #print("update[i] after assigning header is :",temp[i])
                    #print("update[i].header after assigning header is :",temp[i].ptr)
                    #print("update[i].data after assigning header is :",temp[i].data)
                self.height = rlevel 
  
           
            n = self.addStruct(rlevel, data) 
  
             
            for i in range(rlevel+1): 
                n.ptr[i] = temp[i].ptr[i] 
                #print("update[i].forward[i] :",temp[i].ptr[i] )
                temp[i].ptr[i] = n 
                #print("update[",i,"].forward[",i,"] ",temp[i].ptr[i] )
                #print("value is :",temp[i].ptr[i].data)
                #print("value of forward after assignment is :",temp[i].ptr[i])
                #print("value of update key  is :",temp[i].data)
  
            #print("Successfully inserted key {}".format(data)) 
            return temp_level
        #return temp_level    

    
  
```
